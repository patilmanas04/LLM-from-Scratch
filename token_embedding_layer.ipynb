# Implementing the Embedding Layer using the PyTorch

```python
import torch
```

```python
input_ids = torch.tensor([2, 3, 1])
```

```python
vocab_size = 6
output_dim = 3

torch.manual_seed(123)
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
```

```python
print(embedding_layer.weight)
```

<div class="alert-success">
These random weights are further optimized during the LLM training process.
</div>

```python
print(embedding_layer(torch.tensor([3])))
```

# Passing the input_ids and with this the embedding layer will return the weights for the specified input_ids

```python
print(embedding_layer(input_ids))
```
